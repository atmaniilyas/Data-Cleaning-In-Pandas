```python
def clean_last_name(name):
    if isinstance(name, str):  # Only apply regex to strings
        name = re.sub(r"[./_@]", "", name)
        return name.strip()
    else:
        return name  # Leave non-string values (e.g., NaN) as they are
```

```python
df["Last_Name"] = df["Last_Name"].apply(clean_last_name)
```
